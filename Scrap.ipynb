```python
#importo i differenti pacchetti da utilizzare
import urllib.request
from bs4 import BeautifulSoup #importo solo il metodo BeautifulSuop dal pacchetto bs4
from selenium import webdriver
import time
import pandas as pd #rinomino i pacchetti per convenienza
import numpy as np
# specify the url
urlpage = 'http://www.giroditalia.it/it/classifiche/'
# run firefox webdriver from executable path of your choice
driver = webdriver.Firefox() #apro firefox
driver.get(urlpage) #vado sul sito giusto
time.sleep(10)
html = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML") #scarico l'html
soup = BeautifulSoup(html, 'html.parser') #lo rendo compatibile con i metodi(funzioni) di BeautifulSuop
```

```python
My_table = soup.find('table',{'class':'tab_classifiche'}) #trovo la tabella di interesse
dfs = pd.read_html(str(My_table)) #la trasformo in una lista di dataframe
df= dfs[0] #nel primo dataframe c'è la tabella di interesse, negli altri c'è informazione inutile per noi
```
